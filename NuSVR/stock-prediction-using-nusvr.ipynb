```python
#import libraries
from sklearn.svm import NuSVR
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
plt.style.use('fivethirtyeight')
```

```python
#load data
df=pd.read_csv('/kaggle/input/nsei-prev-data/in_nsei_daily.csv',index_col='minute', parse_dates=True)
df.index = df.index.to_pydatetime()
df.shape
```

```python
df = pd.DataFrame(df)
df
```

```python
#prepare data for training SVR models
_df=df.head(len(df)-1)
_df = pd.DataFrame(_df)
_df.shape
```

```python
_df.head
```

```python
_df.index
```

```python
#independent and dependent data
ohl = [[0 for j in range(3)] for i in range(len(_df.index))]
close_price=[]
```

```python
j=0
for i in _df.index:
    ohl[j][0]=int(_df.loc[i].open)
    ohl[j][1]=int(_df.loc[i].high)
    ohl[j][2]=int(_df.loc[i].low)
    j=j+1
for close in _df.close:
    close_price.append(float(close))
```

```python
x_train, x_test, y_train, y_test = train_test_split(ohl, close_price, test_size=0.2)
```

```python
# Create a NuSVR model with default parameters
lin_nu_svr = NuSVR()
poly_nu_svr = NuSVR()
rbf_nu_svr = NuSVR()
```

```python
#3 Support Vector Regressor

#using linear kernel
lin_nu_svr.set_params(kernel='linear', nu=0.5, C=1.0)
print(lin_nu_svr.fit(x_train, y_train))

#using polynomial kernel
poly_nu_svr.set_params(kernel='poly', degree=3, nu=0.5, C=1.0, coef0=1)
print(poly_nu_svr.fit(x_train, y_train))

#using rbf kernel
rbf_nu_svr.set_params(kernel='rbf', nu=0.5, C=1.0, gamma='scale')
print(rbf_nu_svr.fit(x_train, y_train))
```

```python
#plot data vs predictions
plt.figure(figsize=(16,8))
plt.scatter(_df.index,close_price,color='blue',label='Original')
plt.plot(_df.index,rbf_nu_svr.predict(ohl),color='red',label='RBF Model')
plt.plot(_df.index,poly_nu_svr.predict(ohl),color='green',label='Poly Model')
plt.plot(_df.index,lin_nu_svr.predict(ohl),color='yellow',label='Linear Model')
plt.legend()
plt.show()
```

```python
liny_pred=lin_nu_svr.predict(ohl)
polyy_pred=poly_nu_svr.predict(ohl)
rbfy_pred=rbf_nu_svr.predict(ohl)
```

```python
# Calculate the mean squared error of the model
mse = mean_squared_error(close_price, liny_pred)
print("Mean Squared Error RBF:", mse)

mse = mean_squared_error(close_price, polyy_pred)
print("Mean Squared Error RBF:", mse)

mse = mean_squared_error(close_price, rbfy_pred)
print("Mean Squared Error RBF:", mse)
```
