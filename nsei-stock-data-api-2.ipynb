```python
import nsepy as nse
from datetime import date
```

```python
#get history funcction to get Infosys Stock Data
stock_price=nse.get_history(symbol='INFY',index=False, start=date(2023,1,1), end=date(2023,3,15))
stock_price
```

```python
#get history function to get the NIFTY index data
index_price=nse.get_history(symbol='NIFTY',index=True, start=date(2023,1,1), end=date(2023,3,15))
index_price
```

```python
#get history function to get volatility index
vix=nse.get_history(symbol='INDIAVIX',index=True, start=date(2023,1,1), end=date(2023,3,15))
vix
```

```python
#get index p/e information
nifty_pe=nse.get_index_pe_history(symbol='NIFTY', start=date(2023,1,1), end=date(2023,3,15))
nifty_pe
```

```python
#internation amount to indian ruppee coversion
fores=nse.get_rbi_ref_history(start=date(2023,1,1), end=date(2023,3,15))
fores
```

```python
#list as per given date 15 March 2023
bhav_copy=nse.history.get_price_list(dt=date(2023,3,15))
bhav_copy
```

```python
#get list of holidays fall in date range
holidays=nse.live.get_holidays_list(date(2023,1,1),date(2023,3,15))
holidays
```

```python
#get all data related to a stock
quote=nse.live.get_quote(symbol='INFY')
quote
```

```python
import pandas as pd
pd.DataFrame(quote['data']).T
```

```python
#access future data
stock_future=nse.get_history(symbol='SBIN', start=date(2023,1,1), end=date(2023,3,15), index=False, futures=True, expiry_date=date(2023,2,23))
stock_future
```

```python
#access future data
index_future=nse.get_history(symbol='NIFTY', start=date(2023,1,1), end=date(2023,3,15), index=True, futures=True, expiry_date=date(2023,2,23))
index_future
```

```python
#get maturity date, future chain, determine bull and bear
future_chain_table=nse.live.get_futures_chain_table(symbol='SBIN')
future_chain_table
```

```python
#to get expiry date, i.e. last thursday of month
expiry_dates=nse.derivatives.get_expiry_date(year=2023, month=2)
expiry_dates
```

```python
#options data
stock_opt=nse.get_history(symbol='SBIN', start=date(2023,1,1), end=date(2023,3,15), index=False, option_type='CE', strike_price=500, expiry_date=date(2023, 2, 23))
stock_opt
```
